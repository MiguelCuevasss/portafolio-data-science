### ¿Tendría sentido incluir un Análisis Exploratorio de Datos en un pipeline de datos?

Un pipeline tiene:

- Extracción
- Transformación
- Carga 

El EDA aporta:

- Validación de calidad de datos
- Detección de errores
- Identificación de patrones
- Mejora en la toma de decisiones



## ¿Cómo se relaciona con DLM, DL y CRISP-DM?

En CRISP-DM el EDA forma parte de la fase de comprensión de los datos.

Aquí se analizan estructuras, valores faltantes, distribuciones y relaciones.

En DLM El EDA se ubica en la etapa de análisis y uso de datos. Permite entender qué datos son útiles antes de almacenarlos o procesarlos más.

En DL, El EDA es clave antes del entrenamiento porque ayuda a preparar datos, detectar errores y evitar modelos mal entrenados.

```python
import matplotlib.pyplot as plt

import pandas as pd

data = {
    'edad': [20, 25, 30, 35, 40],
    'salario': [1000, 1500, 2000, 2500, 3000],
    'genero': ['M', 'F', 'M', 'F', 'M']
}

df = pd.DataFrame(data)

plt.scatter(df['edad'], df['salario'])
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.title('Relación entre Edad y Salario')
plt.show()
```

```python
import matplotlib.pyplot as plt

import pandas as pd

df = pd.read_csv('archivo.csv')

df['genero'].value_counts().plot(kind='bar')
plt.title('Distribución por Género')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.show()
```

```python
import matplotlib.pyplot as plt

import pandas as pd

df = pd.read_csv('archivo.csv')

plt.hist(df['edad'], bins=10)
plt.title('Distribución de la Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

## ¿Cómo saber cuál es el grafico ideal?

El tipo de gráfico depende del objetivo:

Relación entre variables → Scatter plot
Comparar categorías → Bar plot
Distribución de datos → Histograma
Proporciones → Pie chart

También depende de:

- Tipo de variable (numérica o categórica)
- Cantidad de datos
- Mensaje que se quiere comunicar

##¿Qué es Metadata?

La metadata es información sobre los datos o sobre un paquete.

Ejemplos:

- Versión de un paquete
- Autor
- Licencia
- Dependencias
- Descripción


## Uso de linea de comandos

**pip show pandas**

Output:

- Name: pandas
- Version: 2.x.x
- Summary: Data analysis library
- Author
- Location


**pip list**

Output:

- Un listado de todos los paquetes instalados.


Con la línea de comandos puedo acceder a la metadata de los paquetes instalados, lo cual me permite conocer información relevante como versión, autor, dependencias y ubicación del paquete en el sistema.


## ¿Cómo beneficiaría un buen manejo de metadata a un pipeline de datos?

La metadata ayuda a:

- Entender el origen de los datos
- Controlar versiones
- Mantener consistencia
- Facilitar integración entre sistemas

En un pipeline:

- Permite automatización
- Reduce errores
- Mejora trazabilidad
- Facilita mantenimiento
